# cleaning data for usage

# we will remove all the null values columns 

```python
retials_data.dropna(inplace=True)
```

```python
retials_data.isna
```

```python
# Make the customer ID a string instead of being an integer 
retials_data['CustomerID']= retials_data['CustomerID'].apply(lambda X:'C'+str(int(X)))
```

```python
retials_data.head(15)
```

```python
orders= retials_data[['CustomerID', 'Description', 'InvoiceNo', 'Quantity']]
```

```python
orders.head(10)
```

```python
# Drop descriptions which are not uppercase
orders= orders[orders['Description'].str.upper()==orders['Description']
                          ]
```

```python
orders.head(10)
```

```python
# non-peronalized recommandations
#print out the number of orders that are made in each product 
numbers_orders_perProduct= orders.groupby('Description').agg(orders=('InvoiceNo', 'nunique'), 
                                                                   ).sort_values(by='orders', ascending=False).head(10)
```

```python
numbers_orders_perProduct
```

```python

# print out quantity  of items sold in each product 
quantity_each_sold= orders.groupby('Description').agg(quantity=('Quantity', 'sum')
).sort_values(by='quantity', ascending=False).head(10)
```

```python
quantity_each_sold
```

```python
# create user item interaction with our data 
int_matrix= orders.pivot_table(index='CustomerID', columns=['Description'], values='Quantity', fill_value=0.0)
int_matrix.head(10)
```

```python
# user based remmended system 
active_user= 'C12347'
recommendations= int_matrix.corrwith(int_matrix.loc[active_user], axis=1)
recommendations.dropna(inplace=True)
recommendations
```

```python
# Create DataFram from the corelletion between users
recommandations= pd.DataFrame(recommendations, columns=['Correlation']).reset_index()
recommandations = recommandations.sort_values(by='Correlation', ascending=False)
recommandations.head(10)
```

```python
# find the items that is used the similar user but not the active user
active_user= 'C12347'
similar_user= 'C14326'
for item, quantity in int_matrix.loc[similar_user].items():
    if quantity > 0 and int_matrix.loc[active_user][item]==0:
        print("Name of the item:  {}".format(item)) 
        print("Quantity purchesed by similar user: {}".format(quantity))
        print("Quantity purchesed by active user: {}".format(int_matrix.loc[active_user][item]))
        
        
```
